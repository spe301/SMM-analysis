```python
with open(r'C:\Users\aacjp\OneDrive\Desktop\ssh\Twitter\keys.txt') as f:
    tokens = f.readlines()
```

```python
consumer_key = tokens[0].replace('\n', '')
consumer_secret = tokens[1].replace('\n', '')
access_token = tokens[2].replace('\n', '')
access_token_secret = tokens[3]
import tweepy
import pandas as pd
```

```python
auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)

api = tweepy.API(auth)
```

```python
def TweetDF(topic):
    tweets = api.search(topic)
    c = []
    for i in range(len(tweets)):
        info = tweets[i]._json
        ks = list(info.keys())
        topic_dict = {}
        for k in ks:
            if type(info[k]) != dict:
                topic_dict[k] = info[k]
        row = pd.DataFrame(topic_dict, index=[0])
        c.append(row)
    df = pd.concat(c).reset_index()
    return df

def UserDict(user):
    user = api.get_user(user)._json
    user['url2'] = user['entities']['url']['urls'][0]['url']
    user['expanded'] = user['entities']['url']['urls'][0]['expanded_url']
    user['display'] = user['entities']['url']['urls'][0]['display_url']
    user['created'] = user['status']['created_at']
    user['text'] = user['status']['text']
    user['name'] = user['status']['in_reply_to_screen_name']
    user['sensitive'] = user['status']['possibly_sensitive']
    user_dict = {}
    ks = list(user.keys())
    for k in ks:
        if type(user[k]) != dict:
            user_dict[k] = user[k]
    return user_dict

def UsersDF(users):
    c = []
    for user in users:
        us = pd.DataFrame(UserDict(user), index=[0])
        c.append(us)
    return pd.concat(c).reset_index()

TweetDF('lazer eye surgery')
```
